```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE

anoje = application_df.drop(columns=['EIN', 'NAME'])
anoje
```

```python
anoje.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
#  YOUR CODE GOES HERE

limit = 20110
 
c = anoje['APPLICATION_TYPE'].value_counts()

anoje.loc[anoje['APPLICATION_TYPE'].map(c) < limit, 'APPLICATION_TYPE'] = 'Other'

print(anoje['APPLICATION_TYPE'].value_counts())

```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

cutoff = 294

application_types_to_replace = list(application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < cutoff].index)

application_types_to_replace
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
#  YOUR CODE GOES HERE

 
counting = application_df['CLASSIFICATION'].value_counts()

limit = 1391

classifications_to_replace = counting[counting < limit].index

application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(classifications_to_replace, 'Other')

print(application_df['CLASSIFICATION'].value_counts())
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE

helpful = counting[counting > 1].value_counts

print(helpful)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE


for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE

application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE

from sklearn.model_selection import train_test_split

y = application_df['IS_SUCCESSFUL']
x = application_df.drop('IS_SUCCESSFUL', axis=1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=78)
```

```python

scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE



X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=78)

i = X_train.shape[0]

nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units=8, activation="relu", input_dim=i))

nn.add(tf.keras.layers.Dense(units=5, activation="relu"))

nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE

nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
#  YOUR CODE GOES HERE

fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
#  YOUR CODE GOES HERE

path = 'AlphabetSoupCharity.h5' 

nn.save(path)

print(f"ANOJES MODEL SAVED HERE: {path}")
```
